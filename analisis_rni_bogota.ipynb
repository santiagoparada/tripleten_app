```python
import pandas as pd

df = pd.read_csv('Registro_de_mediciones_de_campos_electromagn_ticos_-_RNI.csv')

print(df.shape)
print(df.columns.tolist())
```

```python
df['ciudad'].value_counts().head(20)
```

```python
bogota = df[df['ciudad'].str.contains('Bogot', case=False, na=False)]

print(bogota.shape)
bogota.sample(10)
```

```python
bogota['ciudad'].value_counts()
```

```python
bogota = bogota.copy()

# tipos de datos correctos
bogota['fecha_medicion'] = pd.to_datetime(bogota['fecha_medicion'], errors='coerce')
bogota['intensidad_campo'] = pd.to_numeric(bogota['intensidad_campo'], errors='coerce')
bogota['máximo_permitido'] = pd.to_numeric(bogota['máximo_permitido'], errors='coerce')
bogota['porcentaje_intensidad_campo'] = pd.to_numeric(bogota['porcentaje_intensidad_campo'], errors='coerce')

# eliminamos nulos
bogota = bogota.dropna()

# eliminamos duplicados
bogota = bogota.drop_duplicates()

# reseteamos el índice
bogota = bogota.reset_index(drop=True)

print(f"Filas después de limpiar: {bogota.shape[0]:,}")
print(f"\nTipos de datos:\n{bogota.dtypes}")
print(f"\nNulos restantes:\n{bogota.isnull().sum()}")
```

```python
bogota['georreferencia'].head(10)
```

```python
bogota.to_csv('bogota_rni_limpio.csv', index=False)
print("Archivo guardado!")
```

```python
bogota[['longitud', 'latitud']] = bogota['georreferencia'].str.extract(r'POINT \(([^ ]+) ([^ ]+)\)').astype(float)

print(bogota[['georreferencia', 'longitud', 'latitud']].head())
```

```python
bogota.to_csv('bogota_rni_limpio.csv', index=False)
print("Archivo actualizado con coordenadas!")
print(bogota.columns.tolist())
```

```python
resumen = bogota.groupby('ubicacion_sonda')['intensidad_campo'].agg(['mean', 'median', 'max', 'min']).reset_index()
resumen = resumen.sort_values('mean', ascending=False)
print(f"Total ubicaciones únicas: {resumen.shape[0]}")
resumen.head(20)
```

```python
ubicaciones_top10 = [
    'Bogota_BATAN',
    'Bogota_CENTRO_COMERCIAL_GRAN_ESTACION',
    'Bogota_AEROCIVIL',
    'Bogota_HOSPITAL_DE_LA_POLICIA',
    'Bogota_EDIFICIO_CLARO',
    'Bogota_DIPOL_(DIRECCION_DE_INTELIGENCIA_POLICIAL)',
    'Bogota_QUIRIGUA',
    'Bogota_LA_ESTRADA',
    'Bogota_SUBA_TIBABUYES',
    'Bogota_CEDRITOS'
]

bogota_top10 = bogota[bogota['ubicacion_sonda'].isin(ubicaciones_top10)]
print(bogota_top10.shape)
bogota_top10['ubicacion_sonda'].value_counts()
```

```python
bogota_top10.to_csv('bogota_top10.csv', index=False)
print("Archivo bogota_top10.csv guardado!")
print(f"Total filas: {bogota_top10.shape[0]:,}")
```

```python
bogota_app = bogota_top10.groupby(
    ['ubicacion_sonda', pd.Grouper(key='fecha_medicion', freq='D')]
).agg(
    intensidad_promedio=('intensidad_campo', 'mean'),
    maximo_permitido=('máximo_permitido', 'first'),
    porcentaje_promedio=('porcentaje_intensidad_campo', 'mean'),
    latitud=('latitud', 'first'),
    longitud=('longitud', 'first')
).reset_index()

bogota_app.to_csv('bogota_app.csv', index=False)
print(f"Filas: {bogota_app.shape[0]:,}")
print(bogota_app.head())
```
